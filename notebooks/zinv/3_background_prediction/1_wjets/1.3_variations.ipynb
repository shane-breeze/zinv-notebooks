### Copy JEC and LEP histograms into central

```python
df_var = pd.concat([df_jec, df_lep], axis=0)
tdf_var = pd.pivot_table(df_var, values=["sum_w", "sum_ww"], index=["varname0", "selection", "parent", "binvar0"], columns=["table"], fill_value=0.)
tdf_var.columns = ["_".join(c) for c in tdf_var.columns]
tdf_var["table"] = "central"
tdf_var = (
    tdf_var.set_index("table", append=True)
    .reorder_levels(["table", "varname0", "selection", "parent", "binvar0"])
)
df = df.merge(tdf_var, how='outer', left_index=True, right_index=True).fillna(0.)
df
```

### Convert W+jets process names

```python
df = dftools.transform.merge(
    df, {
        "WJetsToLNu": [
            "WJetsToEPlusNu", "WJetsToEMinusNu",
            "WJetsToMuPlusNu", "WJetsToMuMinusNu",
            "WJetsToTauLPlusNu", "WJetsToTauLMinusNu", "WJetsToTauHPlusNu", "WJetsToTauHMinusNu",
        ]
    }
)
df
```

### Rebin into final MET binning

```python
metbins = np.array([
    200., 220., 250., 280., 310., 340., 370., 400., 430., 470.,
    510., 550., 590., 640., 690., 740., 790., 840., 900., 960.,
    1020., 1090., 1160., 1250., 1400.,
])
df = dftools.transform.rebin(df, metbins)
```

```python
def draw_variations(data, table="central", varname0="METnoX_pt", selection_parents=[], variations=[], bins=[], ylims=(None, None), scale=1.):
    ncols = len(selection_parents)
    nrows = len(variations)
    fig, axs = plt.subplots(
        figsize=(ncols*scale, nrows*scale),
        nrows=nrows, ncols=ncols,
        sharex=True, sharey=True,
        #gridspec_kw={"hspace": 0.1, "wspace": 0.1},
    )
    
    axs[0,0].set_xlim(bins[0], bins[-1])
    axs[0,0].set_ylim(ylims)
    
    for icol, (selection, parent) in enumerate(selection_parents):
        for irow, variation in enumerate(variations):
            data_plot = data.loc[(table, varname0, selection, parent), ("sum_w", f'sum_w_{variation}Up', f'sum_w_{variation}Down')]
            data_plot = 100*(data_plot.divide(data_plot["sum_w"], axis=0) - 1.)
            
            ax = axs[irow, icol]
            ax.hist(
                bins[:-1], bins=bins, weights=data_plot[f'sum_w_{variation}Up'],
                histtype='step', color='#1f78b4', label=f'{variation}Up',
            )
            ax.hist(
                bins[:-1], bins=bins, weights=data_plot[f'sum_w_{variation}Down'],
                histtype='step', color='#e31a1c', label=f'{variation}Down',
            )
            ax.legend()
            
            if irow == 0:
                ax.set_title(" ".join(selection_parents[icol]))
            
            if irow == nrows-1:
                ax.set_xlabel(r'$\mathcal{U}$ (GeV)')
            if icol == 0:
                ax.set_ylabel(r'Relative variation (\%)')
    return fig, axs
```

```python
draw_variations(
    df,
    selection_parents=[
        ("Monojet", "WJetsToLNu"),
        ("SingleMuon", "WJetsToLNu"),
        ("SingleElectron", "WJetsToLNu"),
    ],
    variations=["jesTotal", "jerSF", "unclust", "muonPtScale", "eleEnergyScale", "tauPtScale", "photonEnergyScale"],
    bins=np.array(list(metbins)+[2*metbins[-1]-metbins[-2]]),
    ylims=(-10., 10.),
    scale=3.,
)
pass
```
